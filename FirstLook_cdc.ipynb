```python
# 1. Basic duplicate statistics
print("=" * 50)
print("STEP 1: BASIC DUPLICATE STATISTICS")
print("=" * 50)

# Identify duplicate rows (all occurrences)
duplicate_mask = df_cdc.duplicated(keep=False)
duplicates = df_cdc[duplicate_mask]
unique_data = df_cdc.drop_duplicates()

# Count statistics
total_rows = len(df_cdc)
unique_rows = len(unique_data)
duplicate_rows = len(df_cdc) - len(unique_data)
duplicate_unique = duplicates.drop_duplicates().shape[0]

print(f"Total rows in dataset: {total_rows:,}")
print(f"Unique rows (patterns): {unique_rows:,}")
print(f"Duplicate rows (subsequent occurrences): {duplicate_rows:,}")
print(f"Percentage of duplicate rows: {duplicate_rows/total_rows*100:.2f}%")
print(f"Number of unique duplicates: {duplicate_unique:,}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Distribution of duplicates across groups
print("\n" + "=" * 50)
print("STEP 3: DISTRIBUTION OF DUPLICATES")
print("=" * 50)

# Let's see if duplicates are more common in certain diabetes status
print("\nDuplicates by diabetes status:")
diabetes_counts = pd.DataFrame({
    'total': df_cdc.groupby('Diabetes_binary').size(),
    'duplicates': duplicates.groupby('Diabetes_binary').size()
})
diabetes_counts['duplicate_rate'] = diabetes_counts['duplicates'] / diabetes_counts['total'] * 100
print(diabetes_counts)

# Check duplication rates across all predictors
demographic_vars = df_cdc.columns.tolist()
for var in demographic_vars:
    print(f"\nDuplication rate by {var}:")
    var_counts = pd.DataFrame({
        'total': df_cdc.groupby(var).size(),
        'duplicates': duplicates.groupby(var).size()
    })
    var_counts['duplicate_rate'] = var_counts['duplicates'] / var_counts['total'] * 100
    print(var_counts.sort_values('duplicate_rate', ascending=False))
    
    # Visualize duplicate rates one by one
    plt.figure(figsize=(10, 6))
    var_counts['duplicate_rate'].sort_values().plot(kind='bar')
    plt.title(f'Duplicate Rate by {var}')
    plt.xlabel(var)
    plt.ylabel('Duplicate Rate (%)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
